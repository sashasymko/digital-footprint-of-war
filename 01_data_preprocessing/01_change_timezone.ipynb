```python
raw_kyiv.mkdir(parents=True, exist_ok=True)
filtered_data.mkdir(parents=True, exist_ok=True)
daily_data.mkdir(parents=True, exist_ok=True)
```

```python
output_dir = Path(raw_kyiv)

files = [
    str(raw_data / "audio_messages_sorted.csv"),
    str(raw_data / "text_messages_sorted.csv"),
    str(raw_data / "comments_sorted.csv"),
    str(raw_data / "posts_sorted.csv"),
    str(raw_data / "reactions_sorted.csv"),
]
```

```python
for path in files:
    df = pd.read_csv(path)
    
    date_col = 'date' if 'date' in df.columns else 'datetime'
    
    df[date_col] = (pd.to_datetime(df[date_col]).dt.tz_localize('UTC').dt.tz_convert('Europe/Kiev').dt.tz_localize(None))
    
    filename = Path(path).name
    df.to_csv(output_dir / filename, index=False)
```
